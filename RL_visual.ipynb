```python
pth='/scratch/kw2815/RL-Finance/checkpoints/2021-09-16/19:48:21'
save_pth = '/scratch/kw2815/RL-Finance/visuals'

# /gae_lambda=0.7,gamma=0.9,net_arch=[16,32,64,32,16]'
```

```python
gamma=[0.9,0.95] 
gae_lambda=[0.7,0.8,0.9] 
net_arch=['[16,32,64,32,16]','[32,64,64,32,16]']
g = gamma[0]
gae = gae_lambda[0]
net = net_arch[0]
```

```python
# df = pd.read_pickle(pth+'/training_env.pkl')
df_loss = pd.read_json(pth+f'/gae_lambda={gae},gamma={g},net_arch={net}'+'/rl_logs.json',lines=True)
df_reward = pd.read_hdf(pth +f'/gae_lambda={gae},gamma={g},net_arch={net}'+ '/reward_history.h5')
```

```python
plot_loss(gae, g, net,'entropy', 'value',pth,save_pth, df=df_loss,cutoff=0.05)
```

```python
plot_reward(gae, g, net, 'revenue', 'reward', save_pth, df=df_reward,freq=1000)
```

```python
for ga, gae, net in itertools.product(gamma, gae_lambda, net_arch):
    plot_loss(gae, ga, net,'entropy', 'value',pth,save_pth)
```

```python
for ga, gae, net in itertools.product(gamma, gae_lambda, net_arch):
    plot_reward(gae, g, net, 'revenue', 'reward', save_pth, pth=pth,freq=1000)
```
